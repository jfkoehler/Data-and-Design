```python
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head(2)
```

```python
penguins['species'].value_counts()
```

```python
X = penguins['species'].value_counts().index
y = penguins['species'].value_counts()
```

```python
plt.bar(X, y)
```

```python
plt.hist(penguins['bill_depth_mm'], bins = 30, 
        alpha = 0.2, edgecolor = 'black');
plt.title('Penguins Bill Depth mm');
```

```python
X = penguins['bill_depth_mm']
y = penguins['bill_length_mm']
plt.scatter(X, y, edgecolor = 'black', alpha = 0.2,
           color = 'salmon');
plt.title('Here is a Scatter Plot of Bills', loc = 'left');
```

### Seaborn

```python
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.countplot(penguins['species'])
```

```python
sns.countplot(penguins['sex'], hue = penguins['species'])
```

```python
sns.boxplot(penguins['species'], penguins['bill_depth_mm'], hue = penguins['sex'])
```

```python
sns.violinplot(penguins['species'], penguins['bill_depth_mm'], hue = penguins['sex'])
```

```python
sns.pairplot(penguins, hue = 'species')
```
