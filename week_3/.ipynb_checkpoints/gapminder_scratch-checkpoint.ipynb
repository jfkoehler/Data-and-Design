```python
def continent_colorer(x):
    '''
    This function takes in a continent
    and returns a color based on continent_dictionary.
    '''
    continent_dictionary =  {'Africa': 'red',
                        'Americas': 'green', 
                        'Asia': 'blue',
                        'Europe': 'yellow',
                        'Oceania': 'purple'} 
    return continent_dictionary[x]
```

```python
gapminder['color'] = gapminder['continent'].apply(continent_colorer)
```

```python
plt.figure(figsize = (15, 6))
plt.scatter(np.log(gapminder['gdpPercap_1952']), 
            gapminder['lifeExp_1952'], 
            c = gapminder['color'],
            s = gapminder['pop_1952']/10**5, alpha = 0.4, marker = 'v')
plt.grid()
plt.title('GDP vs. Life Expectancy in 1952')
```
