```python
#create a scatter plot of economy vs. health
#color points based on region
```

```python
colors = []
for dict_region in happy_2015['Region']:
    if dict_region == 'North America ':
        colors.append('blue')
    elif dict_region == 'Australia and New Zealand':
        colors.append('yellow')
    elif dict_region == 'Eastern Asia ':
        colors.append('green')
    elif dict_region == 'Southern Asia':
        colors.append('darkblue')
    elif dict_region == 'Southeastern Asia':
        colors.append('pink')
    elif dict_region == 'Middle East and Northern Africa':
        colors.append('orange')
    elif dict_region == 'Western Europe':
        colors.append('purple')
    elif dict_region == 'Latin America and Caribbean':
        colors.append('brown')
    elif dict_region == 'Central and Eastern Europe':
        colors.append('turquoise')
    elif dict_region == 'Sub-Saharan Africa ':
        colors.append('coral')
    else:
        colors.append('darkred')
happy_2015['color_graphs'] = colors
```

```python
happy_2015['color_graphs'] = colors
```

```python
plt.scatter(happy_2015['Region'], happy_2015[''])
```

```python
regions = happy_2015.Region.unique()

color = ['blue', 'yellow', 'green', 'darkblue', 'pink', 'orange', 'purple', 'brown', 'turquoise', 'coral', 'darkred']

d = {k:v for k,v in zip(regions, color)}

happy_2015['colors'] = happy_2015.Region.map(d).values
```

```python
happy_2015['colors'].describe()
```

```python
plt.scatter(happy_2015['Economy (GDP per Capita)'], 
            happy_2015['Health (Life Expectancy)'], c = happy_2015['color_graphs'], s = happy_2015['Happiness Score']*5,
           alpha = 0.3, edgecolor = 'black')
```
