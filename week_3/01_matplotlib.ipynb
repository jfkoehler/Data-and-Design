```python
gap_all['continent'].value_counts()
```

```python
def continent_colorer(x):
    color_dict = {'Africa': 'lightblue',
                 'Asia': 'red',
                 'Europe': 'yellow',
                 'Americas': 'green',
                 'Oceania': 'purple'}
    return color_dict[x]
```

```python
gap_all['colors'] = gap_all['continent'].apply(continent_colorer)
```

```python
plt.figure(figsize = (14, 6))
plt.scatter(np.log(gap_all['gdpPercap_1952']), gap_all['lifeExp_1952'], 
            s = gap_all['pop_1952']/10**5.5, 
            c = gap_all['colors'], 
            edgecolor = 'black', alpha = 0.7)
plt.title(f'Wealth vs. Health for 1952', loc = 'left', fontsize = 14)
plt.grid()
```

```python
gdps = []
for column in gap_all.columns:
    if column[0] == 'g':
        gdps.append(column)
```

```python
lifes = []
for column in gap_all.columns:
    if column[0] == 'l':
        lifes.append(column)
```

```python
pops = []
for column in gap_all.columns:
    if column[0] == 'p':
        pops.append(column)
```

```python
gap_all
```

```python
#programmatically extract from columns
gdps = [i for i in gap_all.columns if i.startswith('gdp')]
lifes = [i for i in gap_all.columns if i.startswith('li')]
pops = [i for i in gap_all.columns if i.startswith('po')]
```

```python
#make a list of years
years = [i for i in range(len(pops))]
```

```python
years
```

### The Results

```python
def health_wealth_plotter(year):
    plt.figure(figsize = (14, 6))
    plt.scatter(gap_all[gdps[year]].apply(np.log), gap_all[lifes[year]], 
            s = gap_all[pops[year]]/10**5.5, 
            c = gap_all['colors'], 
            edgecolor = 'black', alpha = 0.7)
    plt.title(f'Wealth vs. Health for {years[year]}', loc = 'left', fontsize = 14)
    plt.grid()
    plt.ylim(20, 90)
    plt.xlim(5, 13)
```

```python
from ipywidgets import interact
import ipywidgets as widgets
```

```python
interact(health_wealth_plotter, 
         year = widgets.IntSlider(0, min = 0, max = len(pops)-1, step = 1, label = years))
```
